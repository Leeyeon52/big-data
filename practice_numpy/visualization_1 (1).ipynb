```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pit
```

```python
pit.rcParams['figure.figsize'] = [10,8]
act = pd.read_csv('OTg6QzA_activities.csv')
act.shape
```

```python
act.head(5)
```

```python
X = act.X.to_numpy()
Y = act.Y.to_numpy()
X
```

```python
Y
```

```python
pit.style.use('grayscale')
pit.figure(facecolor='white')

pit.scatter(X,Y)
pit.show()
```

```python
pit.figure(facecolor='white')
pit.hexbin(X,Y, gridsize=20, cmap=pit.cm.Greys)
cb = pit.colorbar()
cb.set_label('frequency of appearance, gridsize=20')
```

```python
x1 = X[:16570]
x2 = X[16570:16570*2]
x3 = X[16570*2:16750*3]
x4 = X[16570*3:]


y1 = Y[:16570]
y2 = Y[16570:16570*2]
y3 = Y[16570*2:16750*3]
y4 = Y[16570*3:]
```

```python
pit.figure(facecolor='white')

pit.subplot(2,2,1)
pit.hist2d(x1, y1, bins=20, cmap=pit.cm.Greys)
pit.title('1-Q')
pit.colorbar()

pit.subplot(2,2,2)
pit.hist2d(x2, y2, bins=20, cmap=pit.cm.Greys)
pit.title('2-Q')
pit.colorbar()

pit.subplot(2,2,3)
pit.hist2d(x3, y3, bins=20, cmap=pit.cm.Greys)
pit.title('3-Q')
pit.colorbar()

pit.subplot(2,2,4)
pit.hist2d(x4, y4, bins=20, cmap=pit.cm.Greys)
pit.title('4-Q')
pit.colorbar()
```

```python
pit.figure(facecolor='white')

X_dash = X[16570:]
Y_dash = Y[16570:]
pit.hist2d(X_dash, Y_dash, bins=20, cmap=pit.cm.Greys)
cb = pit.colorbar()
cb.set_label('frequency of appearance, bins=20')
```
